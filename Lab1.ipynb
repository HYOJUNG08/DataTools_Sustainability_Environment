# Question1
   
   Create a DataFrame of random numbers with 100 rows and 4 columns. Name the columns a, b, c, and d.

### 1. Make a data frame of random variables and set column names.
---



```python
df=pd.DataFrame(np.random.rand(100,4)) # Create 400 random numbers with hundred rows and four columns
df.columns=['a','b','c','d'] # Name the columns
df
```

# Question 2
Create a scatter plot of column a against b. Set the size of each point on the scatterplot based on a new variable adev, which records the deviations of a from the average value of a.

### Add column 'adev'

* Use 'mean()' and 'std()' method to get the deviations of a from the average value of a.
* Divide by the standard deviation of the df["a"] to do normalization.
* This shows how much the value is far from the average.
* Use 'abs' function to change the negative values to the positive values.
* I am going to use df["adev"] in the next step for the size of the dots. Since size is greater than 0 or equal to 0, size should be a positive number

```python
df["adev"]=abs((df["a"]-df["a"].mean())/df["a"].std())
df
```

### Visualization


```python
# set the figure size
plt.figure(figsize=(16,8))

# x-axis=a, y-axis=b, size = adev * 100, color of the dots = adev
plt.scatter(df.a,df.b,s=df.adev*100, c=df.adev) 

# add x-axis/y-axis label and title
plt.xlabel('a') 
plt.ylabel('b')
plt.title('a scatter plot of column a against b')
```

# Question 3
   
  Graph and examine the correlations between different variables in the EJSCREEN_demo dataset. Comment on the relationship, for example, between education and distance to “treatment storage and disposal facilities” (TSDFs). Is it positive or negative? Are there certain “mass points” or clusters of observations around a certain value (e.g., zero)?

```python
df_ej=pd.read_csv("https://raw.githubusercontent.com/danhammer/envirods/main/data/EJSCREEN_demo.csv")
df_ej
```

### 1. Relationship between the Education and Treatment Storage and Disposal (TSDF) facilities

```python
# set the figure size
plt.figure(figsize=(10,5)) 

# x-axis=LESSHSPCT, y-axis=PTSDF, size = 1
plt.scatter(
    df_ej.LESSHSPCT,
    df_ej.PTSDF,
    s=1
) 

# makes the graph starting from the (0,0)
plt.margins(x=0,y=0) 

# add x-axis/y-axis label and title
plt.xlabel('% individuals age 25 or over with less than high school degree')
plt.ylabel('Proximity to Treatment Storage and Disposal (TSDF) facilities')
plt.title('Relationship between the Education and Treatment Storage and Disposal (TSDF) facilities')
```

* Most of the values cluster around zero to 50 on the graph.
* Besides that values, the graph shows that if people are highly educated, they are further from the treatment storage and disposal facilities compared to people who are not educated.

### 2. Relationship between the income and Treatment Storage and Disposal (TSDF) facilities

```python
# set the figure size
plt.figure(figsize=(10,5)) 

# x-axis=LOWINCPCT, y-axis=PTSDF, size = 1
plt.scatter(
    df_ej.LOWINCPCT,
    df_ej.PTSDF,
    s=1
)

# makes the graph starting from the (0,0)
plt.margins(x=0,y=0)

# add x-axis/y-axis label and title
plt.xlabel('% low income individuals')
plt.ylabel('Proximity to Treatment Storage and Disposal (TSDF) facilities')
plt.title('Relationship between the income and Treatment Storage and Disposal (TSDF) facilities')
```

* Most of the values cluster around zero to 50.
* People with low income are closer to the facilities compared to those with high income.

### 4. Correlations between different variables

* pandas.DataFrame.corr compute pairwise correlation of columns, excluding NA/null values.
* The correlation coefficient value is between -1 and 1, and the absolute value of the correlation coefficient represents a linear relationship.
* If the absolute value of the correlation coefficient is close to 1, it is a perfect correlation, but if it is closer to zero, the weaker the linear relationship.

```python
df_ej.corr()
```



1.   The correlation coefficient between LESSHSPCT and PTSDF is 0.047731.   
Since it is close to zero, it is hard to say that there are linear relationships, but they are in **a positive relationship**. 
2.   The correlation coefficient between LOWINCPCT and PTSDF is 0.004138.   
It is close to zero, so they are not in a linear relationship, but they are in **a positive relationship**.
3.   The correlation coefficient between MINORPCT and PTSDF is 0.127739.   
This one is also close to zero. MINORPCT and PTSDF are not in a linear relationship but **a positive relationship**.
