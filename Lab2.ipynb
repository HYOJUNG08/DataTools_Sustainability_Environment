# Question 3
How many Census Block Groups are there in California (referring to this particular dataset)? How does the average score of “Respiratory Hazard” in California Block Groups compare to the rest of the country? Does this make sense to you?

'nchs' is the new dataframe that gives the data of state and county

```python
nchs = pandas.read_excel("https://github.com/danhammer/envirods/blob/main/data/NCHSURCodes2013.xlsx?raw=true")
nchs = nchs[["FIPS code", "State Abr.", "County name", "2013 code"]]
nchs.columns = ["fips", "state", "county", "classification"]
```

```python
# Associate each of the classifications to one of three categories, rather 
# than one of six - just to simplify our quick analysis
remap_dict = {
    1: "urban",
    2: "suburban",
    3: "suburban",
    4: "rural",
    5: "rural",
    6: "rural"
}

# replace the values of the `classification` column based on the key-value 
# pair in `remap_dict`
nchs = nchs.replace({'classification': remap_dict})
```

```python
#merge the two DataFrames
ejdf = ejdf.merge(nchs, how='left', on='fips')
```

#### 1. 23212 Census Block Groups are in California
Used method 'sum()' to count the number of the row whose value of the 'state' column is CA.

```python
(ejdf["state"]=="CA").sum()
```

#### 2. The average score of“Respiratory Hazard"
Used method "mean()" to get the average. 


```python
#the average score of Respiratory Hazard" in all Block Groups
ejdf["P_RESP_D2"].mean()
```

```python
#Get the data of California from ejdf
ca=ejdf[ejdf["state"]=="CA"]

#the average score of“Respiratory Hazard" in California Block Groups
ca["P_RESP_D2"].mean()
```

```python
#the average score of“Respiratory Hazard" in the rest of the country (california not included)
rest_of_country=ejdf[ejdf["state"]!="CA"]
rest_of_country["P_RESP_D2"].mean()
```

```python
#Sorting the average score of "Respiratory Hazard" in all Block Groups
ejdf.groupby("state")["P_RESP_D2"].mean().sort_values(ascending=False)
```

Comparing the average "Respiratory Hazard" score for California Block Groups to the rest of the country Block Groups, the average score of "Respiratory Hazard" in California Block Groups is greater than the rest of the country. And California is the third highest among all the states. Since California has big cities like San Francisco and LA, it makes sense that the respiratory hazard score for California is higher than other states.

# Question 4
Create single plot with two overlaid histograms. Plot the density of block groups at different respiratory hazard scores (0-100) for California and for the rest of the United States. (Hint: a parameter density=True should be set). The One for California respiratory hazard. Make this look nice by adding a legend and making sure that the histograms are not totally opaque. (Hint: set the parameter alpha to be less than 1.) Comment on the skewness of the distributions.

```python
#using method 'hist()' to make histogram
#alpha gives transparency to the block
#color changes the color of the blocks
#label specifies the each blocks
#density=True makes a histogram into a density histogram

plt.hist(ca["P_RESP_D2"],alpha=0.5,label="ca", bins=100, color="blue",density=True)
plt.hist(rest_of_country["P_RESP_D2"],alpha=0.5,label="rest_of_the_country", bins=100, color="skyblue",density=True)

#the mean and the median of ca["P_RESP_D2"]
plt.axvline(ca["P_RESP_D2"].mean(), color='Red', linestyle='dashed', linewidth=1) 
plt.axvline(ca["P_RESP_D2"].median(), color='Red', linewidth=1)

#the mean and the median of rest_of_country["P_RESP_D2"]
plt.axvline(rest_of_country["P_RESP_D2"].mean(), color='Orange',linestyle='dashed', linewidth=1)
plt.axvline(rest_of_country["P_RESP_D2"].median(), color='Orange', linewidth=1)

#put legend on the upper right part of the graph to describe the elements of the graph
plt.legend(loc="upper right")
```

As you can see from the graph, the mean of the respiratory hazard score of California and the rest of the country is less than the median. So we can say that both of them have negative skewness.

Let's check whether the skewness is negative.   
Used method 'skew()' to get the skewness of each data.

```python
ca["P_RESP_D2"].skew()
```

```python
rest_of_country["P_RESP_D2"].skew()
```

The respiratory hazard score of California has negative skewness, and the respiratory hazard score of the rest of the country also has negative skewness. Both skewnesses mean that the averages of both data are less than median.

According to the graph and skewness we got, the respiratory hazard score for California has negative skewness. We can say that the respiratory hazard score is biased towards 100, and many places in California have a higher respiratory hazard score. But, the respiratory hazard score for the rest of the country, the skewness is negative but close to zero, so we can say that the graph is more likely symmetric. Also, the respiratory hazard score for the rest of the country is relatively evenly distributed.
